```python
#outlier removal
```

```python
def outlier_removal(df,columns):
    for col in columns:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        IQR = q3-q1
        lower = q1-(1.5*IQR)
        upper = q3+(1.5*IQR)
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df  
```

```python
rmf = rmf_df.drop(['customerid'],axis=1)
```

```python
#scaling

from sklearn.preprocessing import MinMaxScaler

```

```python
scaler = MinMaxScaler()
rmf_scaled = scaler.fit_transform(rmf_cleaned[['recency', 'frequency', 'monetary']])
```

```python
# 📌 Why use hue='Frequency'?
# It's optional, but here's why it’s useful:

# Adds a third variable visually:

# The plot already shows Recency (x-axis) and Monetary (y-axis).

# By adding Frequency as hue, you're visualizing all 3 RFM features in one plot.

# Pattern Detection:

# You can see if customers with high Frequency also tend to have low Recency (recent purchases) and high Monetary (spend more).

# This helps validate if there's a natural grouping (i.e., customers that buy often, recently, and spend a lot).

```

```python
rmf_scaled = pd.DataFrame(data=rmf_scaled, columns=['recency', 'frequency', 'monetary'])
```

```python
rmf_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
sse=[]
k_range = range(1,10)
for k in k_range:
    km = KMeans(n_clusters=k,random_state=42)
    km.fit(rmf_scaled[['recency','monetary']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('sse')
plt.plot(k_range,sse)
```

```python
#we can consider no of clusters =2
# model_building
```

```python
km_model = KMeans(n_clusters=2,random_state=42)
km_model
```

```python
y_predicted = km_model.fit_predict(rmf_scaled)
```

```python
y_predicted
```

```python
len(y_predicted)
```

```python
rmf_cleaned['cluster'] = y_predicted
```

```python
rmf_cleaned
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(
    x=rmf_cleaned['recency'], 
    y=rmf_cleaned['monetary'], 
    hue=rmf_cleaned['cluster'], 
    palette='viridis',
    alpha=0.7
)
plt.title('Customer Segments by Recency and Monetary Value')
plt.xlabel('Recency (days)')
plt.ylabel('Monetary Value')
plt.legend(title='Cluster')
plt.show()
```

```python
sns.pairplot(rmf_cleaned, vars=['recency', 'frequency', 'monetary'], hue='cluster', palette='viridis')
plt.show()
```

```python
rmf_cleaned
```

```python
# Compute the average RFM values for each cluster
cluster_profile = rmf_cleaned.groupby('cluster')[['recency', 'frequency', 'monetary']].mean().round(1)
print(cluster_profile)
```

```python
# Cluster 0: Low recency (recent purchases), high frequency, high monetary → Valuable / Loyal Customers

# Cluster 1: High recency (long ago), low frequency, low monetary → Churned / Inactive Customers

```

```python
# Create a mapping manually based on interpretation
segment_map = {
    0: 'Loyal Customers',
    1: 'Churned Customers'
}

# Assign the segment labels
rmf_cleaned['segment'] = rmf_cleaned['cluster'].map(segment_map)
```

```python
rmf_cleaned
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=rmf_cleaned, 
    x='recency', 
    y='monetary', 
    hue='segment', 
    palette='Set1', 
    s=100
)

plt.title("Customer Segments based on Recency and Monetary")
plt.xlabel("Recency (days since last purchase)")
plt.ylabel("Monetary (total spending)")
plt.legend(title='Segment')
plt.grid(True)
plt.show()
```
